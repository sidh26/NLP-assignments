```
import pandas as pd
```

```
df = pd.read_json("/content/train-v2.0.json")
```

```
df1=pd.DataFrame(columns=['title', 'context', 'answer_start', 'text', 'id', 'is_impossible', 'question'])
```

```
l=[]
for row in df['data']:
    title = row.get('title')
    for para in row.get('paragraphs'):
        context = para.get('context')
        for qas in para.get('qas'):
            try:
                answer_start, text = qas.get('answers')[0].get('answer_start'), qas.get('answers')[0].get('text')
            except IndexError:
                answer_start, text = '', ''
            id1, is_impossible, question = qas.get('id'), qas.get('is_impossible'), qas.get('question')
            d={'title':title, 'context':context, 'answer_start': answer_start, 'text': text, 'id':id1, 'is_impossible':is_impossible, 'question':question}
            l.append(d)
df1 = pd.DataFrame(l) 
```

```
import nltk
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')
```

```
import pandas as pd
import re
import numpy as np
from nltk import word_tokenize
from nltk.tokenize import MWETokenizer
from nltk.stem.wordnet import WordNetLemmatizer
from textblob import TextBlob
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.corpus import stopwords
from itertools import combinations
```

```
def preprocess(doc):
    sentences = []
    doc = doc.lower()
    doc = re.sub(r'[^A-Za-z\s]', ' ', doc)
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(doc)
    lem=WordNetLemmatizer()
    t = [lem.lemmatize(w) for w in word_tokens if not w in stop_words]
    t = ' '.join(t)
    return t
```

```
df1['cleantext'] = df1['question'].apply(preprocess)
```

```
df1.head()
```

```
from sklearn.feature_extraction.text import TfidfVectorizer
```

```
vectorizer = TfidfVectorizer(max_features= 8000)
```

```
X = vectorizer.fit_transform(df1['cleantext'])
```

```
temp = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names())
```

```
temp.head()
```

```
quest = []
for i in range(100):
  quest.append(X.toarray()[i].tolist())
quest[0]
```

```
from scipy.spatial import distance
```

```
new_text = "How many concerts did Beyonce perform in?"
nt = preprocess(new_text)
newtext = vectorizer.transform([nt])
new = newtext.toarray().tolist()
new
```

```
dist = []
for j in range(len(quest)):
  dist.append(distance.cosine(new, quest[j]))  
```
